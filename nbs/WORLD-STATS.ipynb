```python
def get_series(df,regions,subject,units,years):
    df_tmp=df[(df['Country'].isin(regions)) & (df['Subject Descriptor']==subject) & (df['Units']==units)]
    cols_to_keep=['Country']+ years
    df_tmp=df_tmp[cols_to_keep]
    df_tmp=df_tmp.rename({'Country':''},axis=1)
    df_tmp=df_tmp.set_index('')
    df_tmp=df_tmp.T.reset_index()
    df_tmp=df_tmp.rename({'index':'Year'},axis=1)
    return df_tmp
```

```python
regions=['World']
subject='Gross domestic product, current prices'
units='U.S. dollars'
years=np.arange(1980,2028).tolist()
df_tmp=get_series(df_world,regions,subject,units,years)
# df_tmp=df_world[(df_world['Country Group Name'].isin(country)) & (df_world['Subject Descriptor']==subject) & (df_world['Units']==units)]
# estimate_year=int(df_tmp.loc[df_tmp.index[0],'Estimates Start After'])
df_tmp.tail()
```

```python
regions=['United States','China','India','France','Japan','Canada','Brasil','Germany','Mexico','United Kingdom']
subject='Gross domestic product, current prices'
units='U.S. dollars'
year=2023
years=np.arange(1980,2028).tolist()
# df_tmp=get_series(df_imf,regions,subject,units,years)
df_tmp=df_imf[(df_imf['Country'].isin(regions)) & (df_imf['Subject Descriptor']==subject) & (df_imf['Units']==units)]
cols_to_keep=['ISO',year]
df_tmp=df_tmp[cols_to_keep]
df_tmp=df_tmp.rename({year:str(year)},axis=1)
df_tmp.sort_values(by=str(year),ascending=False,inplace=True)
print(df_tmp.shape)
df_tmp
```

```python
df_tmp['CumGDP']=df_tmp[str(year)].cumsum()
df_tmp
```

```python
base = alt.Chart(df_tmp).encode(
    alt.Theta("2023:Q").stack(True),
    alt.Color("ISO:N").legend(None)
).interactive()

pie = base.mark_arc(outerRadius=120)
text = base.mark_text(radius=140, size=20).encode(text="ISO:N")

pie + text
```
