# Impact of phase on a signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=np.arange(0,170,1)
len(n)
```

```python
w=(2*np.pi)/40
w
```

## Zero phase

```python
x=(1/2)*np.sin(w*n)+np.cos(2*w*n)
plt.figure(figsize=(12,8))
plt.plot(n,x)
```

We can call this signal zero phase signal as the phase associated to each sinusoid is zero.

## Linear Phase

```python
#Let's add a phase term to the above signal.
theta=8*np.pi/5
```

```python
x_linphase=(1/2)*np.sin(w*n+theta)+np.cos(2*(w*n+theta))
plt.figure(figsize=(12,8))
plt.plot(n,x_linphase)
```

We see that when we add a phase term to the sinusoidal components that is proportional to the freq of the signal, the net effect in the time domain is the shift of the signal. The shape of the signal is unchanged.

Now let's add a phase that is non-proportional to the frequency,

## Non-linear phase

```python
x_nonlinphase=(1/2)*np.sin(w*n)+np.cos(2*(w*n+theta))
plt.figure(figsize=(12,8))
plt.plot(n,x_nonlinphase)
```

Now, the shape of the signal changes.
